```python
import gnwrapper
import gym

#Breakout-v0
env = gnwrapper.Animation(gym.make('Breakout-v0')) # Start Xvfb

o = env.reset()

for _ in range(100):
    o, r, d, i = env.step(env.action_space.sample()) # Take action from DNN
    env.render() # Here, clear old display and redraw new display
    if d:
        env.reset()
```
